```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def test_distributions(data, dist_names=["norm", "expon", "triang"]):
    """
    Test multiple distributions and compute the Kolmogorov-Smirnov test for goodness-of-fit.
    
    Parameters:
        data (array-like): Sample data.
        dist_names (list): List of distribution names to test.
    
    Returns:
        dict: Dictionary containing parameters, KS statistic, and p-value for each distribution.
    """
    results = {}
    for dist_name in dist_names:
        dist = getattr(stats, dist_name)
        params = dist.fit(data)  # Fit the distribution to the data
        ks_stat, p_value = stats.kstest(data, dist_name, args=params)
        results[dist_name] = {"params": params, "ks_stat": ks_stat, "p_value": p_value}
    return results

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a given dataset.
    
    Parameters:
        data (array-like): Sample data.
        confidence (float): Confidence level (default: 0.95 for 95% CI).
    
    Returns:
        tuple: (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2., n-1)
    
    return mean - margin, mean + margin

# Load dataset (assuming it is structured with process times in rows, and samples in columns)
file_path = "L.csv"
df = pd.read_csv(file_path)

# Extract service times for a specific process (e.g., Initial Phase)
initial_phase_times = df.iloc[0, 1:].astype(float).values  # Adjust row index accordingly

# Run distribution testing
distribution_results = test_distributions(initial_phase_times)

# Print results
for dist, res in distribution_results.items():
    print(f"\nDistribution: {dist}")
    print(f"Parameters: {res['params']}")
    print(f"KS Statistic: {res['ks_stat']}")
    print(f"P-Value: {res['p_value']}")

# Compute confidence interval for the selected process times
ci_lower, ci_upper = confidence_interval(initial_phase_times)
print(f"\n95% Confidence Interval for Initial Phase Processing Time: ({ci_lower:.4f}, {ci_upper:.4f})")

# Plot histogram with fitted distributions
plt.figure(figsize=(8,5))
plt.hist(initial_phase_times, bins=20, density=True, alpha=0.6, label="Observed Data")
x = np.linspace(min(initial_phase_times), max(initial_phase_times), 100)

# Overlay fitted distributions
for dist, res in distribution_results.items():
    dist_obj = getattr(stats, dist)
    plt.plot(x, dist_obj.pdf(x, *res['params']), label=f"{dist.capitalize()} Fit")

plt.xlabel("Processing Time (Minutes)")
plt.ylabel("Density")
plt.title("Fitting Distributions and Confidence Interval for Process Times")
plt.legend()
plt.show()
```
